```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
import pandas as pd

data = pd.read_csv("data.csv", encoding="latin1")   # or "ISO-8859-1"

x_values = data.iloc[:, 0].values
n = len(x_values)

print("Total samples:", n)
```

```python
# Convert to numeric (force non-numeric values to NaN)
x_values = pd.to_numeric(x_values, errors="coerce")

# Remove NaN values using NumPy
x_values = x_values[~np.isnan(x_values)]

print("Valid numeric samples:", len(x_values))

```

```python
import matplotlib.pyplot as plt

plt.hist(x_values, bins=25, density=True, alpha=0.7)
plt.title("Histogram of Cleaned Data")
plt.xlabel("Sample Values")
plt.ylabel("Probability Density")
plt.grid(True)
plt.show()
```

```python
mu_hat = np.mean(x_values)
sigma_hat = np.sqrt(np.mean((x_values - mu_hat)**2))

print("MLE Estimated Mean (μ):", mu_hat)
print("MLE Estimated Standard Deviation (σ):", sigma_hat)
```

```python
from scipy.stats import norm

x_axis = np.linspace(np.min(x_values), np.max(x_values), 300)
estimated_pdf = norm.pdf(x_axis, mu_hat, sigma_hat)
```

```python
plt.hist(x_values, bins=25, density=True, alpha=0.6, label="Observed Data")
plt.plot(x_axis, estimated_pdf, linewidth=2, label="MLE Estimated PDF")

plt.xlabel("x")
plt.ylabel("Density")
plt.title("PDF Estimation using MLE")
plt.legend()
plt.grid(True)
plt.show()
```

```python
log_likelihood = np.sum(np.log(norm.pdf(x_values, mu_hat, sigma_hat)))
print("Log Likelihood:", log_likelihood)
```
